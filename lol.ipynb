```python
# Dar uma olhada como as variaves se correlacionam
# Da pra notar que o numero de abates de um time tem 100 de correlação com o numero de mortes do outro time (A matriz ta funcionando)

corr =  df_clean.corr().round(2)

mask = np.triu(np.ones_like(corr, dtype=bool))

f, ax = plt.subplots(figsize=(16, len(df_clean.columns)))
sns.heatmap(data = corr, mask = mask,annot=True,cmap='coolwarm', vmin=-1, vmax=1, center = 0, square = True, linewidths=0.6, cbar_kws={"shrink": .5})
ax.set_title('Matriz de correlação', fontsize = 25)
plt.show()
```

# Machine Learning

```python
# Vou tentar limpar mais ainda o dataframe, deu para notar no heatmap feito anteriormente que tem alguns features...
# Que não estao com bons valores de correlação ao target (blueWins)

columns = []

sel = df_clean.corr()['blueWins']
for i in range(sel.shape[0]):
    if sel[i] > 0.2:
        columns.append(sel.index[i])
    elif sel[i] < -0.2:
        columns.append(sel.index[i])
        
        
        
columns
df_ml = df_clean[columns]
print(df_ml.columns)
```

#### Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.model_selection import GridSearchCV


X = df_ml.drop(columns = 'blueWins')
y = df_ml['blueWins']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)
```

```python
# Eu esqueci os parametros pro tunning, so pra lembrar mesmo
rf = RandomForestClassifier()
rf.get_params()
```

```python
#rf = RandomForestClassifier(n_estimators=400, min_samples_leaf=0.12, random_state=42)


# Gridsearch
grid = {'n_estimators': [100,200,300,400,500], 'max_depth': [2,5,10], 'min_samples_leaf': [0.1, 0.2]}
grid_rf = GridSearchCV(estimator = rf, param_grid = grid, cv = 5, n_jobs=-1)

# Fit 'rf' to the training set
grid_rf.fit(X_train, y_train)

# Predict the test set labels 'y_pred'
y_pred = grid_rf.predict(X_test)
```

```python
accuracy_score(y_test, y_pred)
```

```python
grid_rf.best_estimator_
```

```python
best_rf = grid_rf.best_estimator_

best_rf.fit(X_train, y_train)
y_pred = best_rf.predict(X_test)

accuracy_score(y_test, y_pred)
```

```python
importances_rf = pd.Series(grid_rf.best_estimator_.feature_importances_, index = X.columns)

sorted_importances_rf = importances_rf.sort_values()

# Make a horizontal bar plot
sorted_importances_rf.plot(kind='barh', color='lightgreen')
plt.show()
```

#### Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
dt.get_params()
```

```python
grid = {'max_depth': [3, 4, 5, 6], 'min_samples_leaf': [0.04, 0.06, 0.08], 'max_features': [0.2, 0.4, 0.6, 0.8]}
grid_dt = GridSearchCV(estimator = dt, param_grid = grid,  scoring='accuracy', cv=10, n_jobs=-1)

grid_dt.fit(X_train, y_train)
y_pred = grid_dt.predict(X_test)
accuracy_score(y_test, y_pred)
```

```python
best_dt = grid_dt.best_estimator_

best_dt.fit(X_train, y_train)
y_pred = best_dt.predict(X_test)

accuracy_score(y_test, y_pred)
```

#### Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB

nb = GaussianNB()
nb.fit(X_train, y_train)

pred_nb = nb.predict(X_test)

# get the accuracy score
acc_nb = accuracy_score(pred_nb, y_test)
print(acc_nb)
```

#### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(X_train, y_train)

# get accuracy score
pred_lr = lr.predict(X_test)
acc_lr = accuracy_score(pred_lr, y_test)
print(acc_lr)
```
